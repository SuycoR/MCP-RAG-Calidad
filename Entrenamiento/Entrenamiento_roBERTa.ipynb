```python
# ==============================================
# 🧠 ENTRENAMIENTO DE ROBERTA CON CLASS WEIGHTS Y F1 MACRO
# ==============================================
# Este notebook:
# - Usa class weights en lugar de oversampling
# - Optimiza F1-macro en lugar de F1-weighted
# - Implementa CustomTrainer con weighted loss

# ==============================================
# 🔧 1. VERIFICACIÓN Y ACTUALIZACIÓN DE LIBRERÍAS
# ==============================================
import importlib, sys, subprocess

def instalar_paquete(paquete, version=None):
    if version:
        subprocess.check_call([sys.executable, "-m", "pip", "install", f"{paquete}=={version}"])
    else:
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-U", paquete])

paquetes_requeridos = {
    "transformers": "4.44.2",
    "datasets": "3.0.1",
    "torch": None,
    "accelerate": "0.34.2",
    "scikit-learn": None,
    "matplotlib": None,
    "seaborn": None
}

for paquete, version in paquetes_requeridos.items():
    try:
        mod = importlib.import_module(paquete)
        print(f"✅ {paquete} ya instalado (versión {mod.__version__})")
    except Exception:
        print(f"⬇️ Instalando {paquete}...")
        instalar_paquete(paquete, version)

# ==============================================
# 📦 2. IMPORTAR LIBRERÍAS
# ==============================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import torch
import torch.nn as nn
from datasets import load_dataset, Dataset
from transformers import RobertaTokenizer, RobertaForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
from collections import Counter
from google.colab import files
import wandb

# 🔐 CONFIGURAR WANDB
os.environ["WANDB_API_KEY"] = "5b52cc96b7af52e939c73bf8f1fcf8c7895e36a1"
wandb.login(key="5b52cc96b7af52e939c73bf8f1fcf8c7895e36a1")
print("✅ Wandb configurado correctamente")

# ==============================================
# 📂 3. CARGAR DATASET
# ==============================================
print("\n📤 Sube tu archivo 'TrainingData_traducido.csv'")
uploaded = files.upload()

df = pd.read_csv("TrainingData_traducido.csv", encoding="latin1")
print(f"✅ Dataset cargado con {len(df)} filas")
print(df.head())

# ==============================================
# 🔧 4. PROCESAMIENTO DE ETIQUETAS (SIN BALANCEO)
# ==============================================
df = df[["new_message1", "label"]].dropna()
df = df.rename(columns={"new_message1": "text"})

print(f"\n📊 Distribución ORIGINAL de clases:")
print("   Significado de las clases:")
print("   - Clase 0: Documentado (contiene Why AND What)")
print("   - Clase 1: Ambiguo (no contiene Why NOR What)")
print("   - Clase 2: Vago - No What (falta el qué se hizo)")
print("   - Clase 3: Vago - No Why (falta el por qué se hizo)")
print("\n   Distribución:")
distribucion_original = df["label"].value_counts().sort_index()
print(distribucion_original)
print(f"\n   Total de ejemplos: {len(df)}")
for clase, count in distribucion_original.items():
    porcentaje = (count / len(df)) * 100
    print(f"   Clase {clase}: {count} ejemplos ({porcentaje:.1f}%)")

# ⚖️ CALCULAR CLASS WEIGHTS
class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(df['label']),
    y=df['label']
)
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float32)

print(f"\n⚖️ Class Weights calculados (estrategia 'balanced'):")
for i, weight in enumerate(class_weights):
    print(f"   Clase {i}: {weight:.4f}")

# División entrenamiento / prueba (80/20)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df["label"])

print(f"\n✅ División realizada (80% entrenamiento / 20% prueba):")
print(f"   - Entrenamiento: {len(train_df)} ejemplos (~80%)")
print(f"   - Prueba: {len(test_df)} ejemplos (~20%)")
print(f"\n   Distribución en entrenamiento:")
print(train_df["label"].value_counts().sort_index())
print(f"\n   Distribución en prueba:")
print(test_df["label"].value_counts().sort_index())

# Crear datasets de Hugging Face
train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(test_df)

# ==============================================
# 🔤 5. TOKENIZACIÓN
# ==============================================
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_test = test_dataset.map(tokenize_function, batched=True)

# ==============================================
# 🤖 6. CUSTOM TRAINER CON CLASS WEIGHTS
# ==============================================
class WeightedLossTrainer(Trainer):
    def __init__(self, *args, class_weights=None, **kwargs):
        super().__init__(*args, **kwargs)
        self.class_weights = class_weights

    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        """
        Calcula la loss con class weights aplicados.
        Compatible con transformers >= 4.44.0
        """
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits

        # Aplicar class weights a la loss
        loss_fct = nn.CrossEntropyLoss(weight=self.class_weights.to(logits.device))
        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))

        return (loss, outputs) if return_outputs else loss

# Cargar modelo
model = RobertaForSequenceClassification.from_pretrained("roberta-base", num_labels=4)
print("✅ Modelo cargado con 4 clases de salida")

# ==============================================
# 📊 7. MÉTRICAS CON F1 MACRO
# ==============================================
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)

    # F1 macro (promedio simple de F1 por clase)
    precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(
        labels, preds, average="macro"
    )

    # F1 weighted (promedio ponderado por soporte)
    precision_weighted, recall_weighted, f1_weighted, _ = precision_recall_fscore_support(
        labels, preds, average="weighted"
    )

    # Accuracy
    acc = accuracy_score(labels, preds)

    # F1 por clase individual
    precision_per_class, recall_per_class, f1_per_class, _ = precision_recall_fscore_support(
        labels, preds, average=None
    )

    metrics = {
        "accuracy": acc,
        "f1_macro": f1_macro,  # ✅ Métrica principal
        "precision_macro": precision_macro,
        "recall_macro": recall_macro,
        "f1_weighted": f1_weighted,
        "precision_weighted": precision_weighted,
        "recall_weighted": recall_weighted,
    }

    # Agregar F1 por clase
    for i in range(len(f1_per_class)):
        metrics[f"f1_class_{i}"] = f1_per_class[i]

    return metrics

# ==============================================
# ⚙️ 8. CONFIGURACIÓN DEL ENTRENAMIENTO
# ==============================================
wandb.init(
    project="commit-classification-roberta_V2",
    name="roberta-4clases-classweights-f1macro",
    config={
        "model": "roberta-base",
        "num_labels": 4,
        "learning_rate": 2e-5,
        "batch_size": 16,
        "epochs": 5,
        "max_length": 128,
        "balancing_strategy": "class_weights",
        "optimization_metric": "f1_macro",
        "class_weights": class_weights.tolist(),
        "weight_decay": 0.01
    }
)

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=50,
    load_best_model_at_end=True,
    metric_for_best_model="f1_macro",  # ✅ Optimizar F1 macro
    greater_is_better=True,
    dataloader_pin_memory=False,
    fp16=True,
    report_to="wandb",
    warmup_steps=100,
    save_total_limit=2
)

print("\n⚙️ Configuración del entrenamiento:")
print(f"   - Estrategia: Class Weights (sin oversampling)")
print(f"   - Métrica objetivo: F1-macro")
print(f"   - Learning rate: 2e-5")
print(f"   - Epochs: 5")
print(f"   - Batch size: 16")
print(f"   - Class weights aplicados en la loss function")

# ==============================================
# 🚀 9. ENTRENAMIENTO DEL MODELO
# ==============================================
trainer = WeightedLossTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
    class_weights=class_weights_tensor
)

print("\n🚀 Iniciando entrenamiento con class weights...\n")
print("   ⏱️ Tiempo estimado: ~15-20 minutos con GPU")
print("   📊 Sigue el progreso en tu dashboard de Wandb\n")
trainer.train()

# ==============================================
# 🧾 10. EVALUACIÓN FINAL
# ==============================================
metrics = trainer.evaluate()
print("\n📈 Resultados de evaluación:")
print(f"\n   🎯 Métricas principales:")
print(f"   - Accuracy: {metrics['eval_accuracy']:.4f}")
print(f"   - F1-macro: {metrics['eval_f1_macro']:.4f} ⭐")
print(f"   - F1-weighted: {metrics['eval_f1_weighted']:.4f}")
print(f"   - Precision-macro: {metrics['eval_precision_macro']:.4f}")
print(f"   - Recall-macro: {metrics['eval_recall_macro']:.4f}")

print(f"\n   📊 F1-score por clase:")
for i in range(4):
    print(f"   - Clase {i}: {metrics[f'eval_f1_class_{i}']:.4f}")

# ==============================================
# 📉 11. VISUALIZACIÓN DE MÉTRICAS
# ==============================================
preds_output = trainer.predict(tokenized_test)
y_true = preds_output.label_ids
y_pred = preds_output.predictions.argmax(-1)

# Matriz de confusión
cm = confusion_matrix(y_true, y_pred, labels=[0, 1, 2, 3])

wandb.log({
    "confusion_matrix": wandb.plot.confusion_matrix(
        probs=None,
        y_true=y_true,
        preds=y_pred,
        class_names=["0-Doc", "1-Amb", "2-NoWhat", "3-NoWhy"]
    )
})

plt.figure(figsize=(8,7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=[0, 1, 2, 3], yticklabels=[0, 1, 2, 3])
plt.xlabel("Predicciones")
plt.ylabel("Etiquetas reales")
plt.title("Matriz de Confusión - RoBERTa con Class Weights")
plt.tight_layout()
plt.show()

# Reporte detallado
labels_names = [
    "Documentado (Why+What)",
    "Ambiguo (No Why, No What)",
    "Vago - No What",
    "Vago - No Why"
]
report = classification_report(y_true, y_pred, target_names=labels_names, output_dict=True)
report_df = pd.DataFrame(report).transpose()

print("\n📋 Reporte detallado por clase:")
print(report_df)

# Gráfico comparativo F1-macro vs F1-weighted vs F1 por clase
plt.figure(figsize=(12,6))

# F1 por clase
clases = [0, 1, 2, 3]
f1_scores = [report[labels_names[i]]["f1-score"] for i in clases]
colors = ['#2E7D32', '#F57C00', '#C62828', '#6A1B9A']
nombres_cortos = ['Doc\n(0)', 'Amb\n(1)', 'NoWhat\n(2)', 'NoWhy\n(3)']

x_pos = np.arange(len(clases))
bars = plt.bar(x_pos, f1_scores, color=colors, alpha=0.7, edgecolor='black', label='F1 por clase')

# Línea de F1-macro
f1_macro_value = np.mean(f1_scores)
plt.axhline(y=f1_macro_value, color='red', linestyle='--', linewidth=2, label=f'F1-macro: {f1_macro_value:.3f}')

# Línea de F1-weighted
support = [report[labels_names[i]]["support"] for i in clases]
f1_weighted_value = np.average(f1_scores, weights=support)
plt.axhline(y=f1_weighted_value, color='blue', linestyle='--', linewidth=2, label=f'F1-weighted: {f1_weighted_value:.3f}')

plt.ylim(0, 1)
plt.xlabel("Clase")
plt.ylabel("F1-score")
plt.title("📊 Comparación F1: Por Clase vs Macro vs Weighted")
plt.xticks(x_pos, nombres_cortos)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Log métricas individuales a wandb
for i, nombre in enumerate(labels_names):
    wandb.log({
        f"f1_score_clase_{i}": f1_scores[i],
        f"precision_clase_{i}": report[nombre]["precision"],
        f"recall_clase_{i}": report[nombre]["recall"]
    })

# Análisis de errores
print("\n📊 Análisis de predicciones por clase:")
for i in range(4):
    mascara_real = y_true == i
    total_clase = np.sum(mascara_real)
    correctos = np.sum((y_true == i) & (y_pred == i))
    incorrectos = total_clase - correctos

    if total_clase > 0:
        precision_clase = (correctos / total_clase) * 100
        print(f"\nClase {i} - {labels_names[i]}:")
        print(f"   Total: {total_clase} ejemplos")
        print(f"   Correctos: {correctos} ({precision_clase:.1f}%)")
        print(f"   Incorrectos: {incorrectos} ({100-precision_clase:.1f}%)")
        print(f"   Class weight aplicado: {class_weights[i]:.4f}")

        if incorrectos > 0:
            print(f"   Confusiones:")
            for j in range(4):
                if i != j:
                    confusiones = np.sum((y_true == i) & (y_pred == j))
                    if confusiones > 0:
                        print(f"      - {confusiones} casos → Clase {j} ({labels_names[j]})")

# ==============================================
# 💾 12. GUARDAR Y DESCARGAR MODELO
# ==============================================
model.save_pretrained("./modelo_roberta_classweights_f1macro")
tokenizer.save_pretrained("./modelo_roberta_classweights_f1macro")

# Guardar configuración
config_info = {
    "num_labels": 4,
    "label_mapping": {
        "0": "Documentado (Why + What)",
        "1": "Ambiguo (Neither Why nor What)",
        "2": "No What",
        "3": "No Why"
    },
    "class_weights": class_weights.tolist(),
    "optimization_metric": "f1_macro",
    "training_strategy": "class_weights_in_loss"
}

import json
with open("./modelo_roberta_classweights_f1macro/model_config.json", "w") as f:
    json.dump(config_info, f, indent=2)

print("\n✅ Modelo guardado en './modelo_roberta_classweights_f1macro'")

# Comprimir y descargar
!zip -r modelo_roberta_classweights_f1macro.zip modelo_roberta_classweights_f1macro
files.download("modelo_roberta_classweights_f1macro.zip")

# Guardar métricas
metricas_por_clase = pd.DataFrame({
    "Clase": [0, 1, 2, 3],
    "Descripcion": labels_names,
    "Precision": [report[labels_names[i]]["precision"] for i in range(4)],
    "Recall": [report[labels_names[i]]["recall"] for i in range(4)],
    "F1-score": [report[labels_names[i]]["f1-score"] for i in range(4)],
    "Support": [report[labels_names[i]]["support"] for i in range(4)],
    "Class_Weight": class_weights.tolist()
})

metricas_por_clase.to_csv("metricas_classweights_f1macro.csv", index=False)
files.download("metricas_classweights_f1macro.csv")

# Resumen final
print("\n🎉 ¡Proceso completado!")
print(f"\n📊 Resumen de métricas:")
print(f"   - Accuracy: {metrics['eval_accuracy']:.4f}")
print(f"   - F1-macro (objetivo): {metrics['eval_f1_macro']:.4f} ⭐")
print(f"   - F1-weighted: {metrics['eval_f1_weighted']:.4f}")
print(f"\n⚖️ Estrategia utilizada:")
print(f"   - Class Weights en loss function")
print(f"   - Sin oversampling artificial")
print(f"   - Optimización basada en F1-macro")

wandb.finish()
```

```python
# ==============================================
# 🧠 ENTRENAMIENTO DE ROBERTA CON CLASS WEIGHTS Y F1 MACRO
# ==============================================
# Este notebook:
# - Usa class weights en lugar de oversampling
# - Optimiza F1-macro en lugar de F1-weighted
# - Implementa CustomTrainer con weighted loss

# ==============================================
# 🔧 1. VERIFICACIÓN Y ACTUALIZACIÓN DE LIBRERÍAS
# ==============================================
import importlib, sys, subprocess

def instalar_paquete(paquete, version=None):
    if version:
        subprocess.check_call([sys.executable, "-m", "pip", "install", f"{paquete}=={version}"])
    else:
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-U", paquete])

paquetes_requeridos = {
    "transformers": "4.44.2",
    "datasets": "3.0.1",
    "torch": None,
    "accelerate": "0.34.2",
    "scikit-learn": None,
    "matplotlib": None,
    "seaborn": None
}

for paquete, version in paquetes_requeridos.items():
    try:
        mod = importlib.import_module(paquete)
        print(f"✅ {paquete} ya instalado (versión {mod.__version__})")
    except Exception:
        print(f"⬇️ Instalando {paquete}...")
        instalar_paquete(paquete, version)

# ==============================================
# 📦 2. IMPORTAR LIBRERÍAS
# ==============================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import torch
import torch.nn as nn
from datasets import load_dataset, Dataset
from transformers import RobertaTokenizer, RobertaForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
from collections import Counter
from google.colab import files
import wandb

# 🔐 CONFIGURAR WANDB
os.environ["WANDB_API_KEY"] = "5b52cc96b7af52e939c73bf8f1fcf8c7895e36a1"
wandb.login(key="5b52cc96b7af52e939c73bf8f1fcf8c7895e36a1")
print("✅ Wandb configurado correctamente")

# ==============================================
# 📂 3. CARGAR DATASET
# ==============================================
print("\n📤 Sube tu archivo 'TrainingData_traducido.csv'")
uploaded = files.upload()

df = pd.read_csv("TrainingData_traducido.csv", encoding="latin1")
print(f"✅ Dataset cargado con {len(df)} filas")
print(df.head())

# ==============================================
# 🔧 4. PROCESAMIENTO DE ETIQUETAS (SIN BALANCEO)
# ==============================================
df = df[["new_message1", "label"]].dropna()
df = df.rename(columns={"new_message1": "text"})

print(f"\n📊 Distribución ORIGINAL de clases:")
print("   Significado de las clases:")
print("   - Clase 0: Documentado (contiene Why AND What)")
print("   - Clase 1: Ambiguo (no contiene Why NOR What)")
print("   - Clase 2: Vago - No What (falta el qué se hizo)")
print("   - Clase 3: Vago - No Why (falta el por qué se hizo)")
print("\n   Distribución:")
distribucion_original = df["label"].value_counts().sort_index()
print(distribucion_original)
print(f"\n   Total de ejemplos: {len(df)}")
for clase, count in distribucion_original.items():
    porcentaje = (count / len(df)) * 100
    print(f"   Clase {clase}: {count} ejemplos ({porcentaje:.1f}%)")

# ⚖️ CALCULAR CLASS WEIGHTS
class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(df['label']),
    y=df['label']
)
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float32)

print(f"\n⚖️ Class Weights calculados (estrategia 'balanced'):")
for i, weight in enumerate(class_weights):
    print(f"   Clase {i}: {weight:.4f}")

# División entrenamiento / prueba (80/20)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df["label"])

print(f"\n✅ División realizada (80% entrenamiento / 20% prueba):")
print(f"   - Entrenamiento: {len(train_df)} ejemplos (~80%)")
print(f"   - Prueba: {len(test_df)} ejemplos (~20%)")
print(f"\n   Distribución en entrenamiento:")
print(train_df["label"].value_counts().sort_index())
print(f"\n   Distribución en prueba:")
print(test_df["label"].value_counts().sort_index())

# Crear datasets de Hugging Face
train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(test_df)

# ==============================================
# 🔤 5. TOKENIZACIÓN
# ==============================================
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_test = test_dataset.map(tokenize_function, batched=True)

# ==============================================
# 🤖 6. CUSTOM TRAINER CON CLASS WEIGHTS
# ==============================================
class WeightedLossTrainer(Trainer):
    def __init__(self, *args, class_weights=None, **kwargs):
        super().__init__(*args, **kwargs)
        self.class_weights = class_weights

    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        """
        Calcula la loss con class weights aplicados.
        Compatible con transformers >= 4.44.0
        """
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits

        # Aplicar class weights a la loss
        loss_fct = nn.CrossEntropyLoss(weight=self.class_weights.to(logits.device))
        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))

        return (loss, outputs) if return_outputs else loss

# Cargar modelo
model = RobertaForSequenceClassification.from_pretrained("roberta-base", num_labels=4)
print("✅ Modelo cargado con 4 clases de salida")

# ==============================================
# 📊 7. MÉTRICAS CON F1 MACRO
# ==============================================
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)

    # F1 macro (promedio simple de F1 por clase)
    precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(
        labels, preds, average="macro"
    )

    # F1 weighted (promedio ponderado por soporte)
    precision_weighted, recall_weighted, f1_weighted, _ = precision_recall_fscore_support(
        labels, preds, average="weighted"
    )

    # Accuracy
    acc = accuracy_score(labels, preds)

    # F1 por clase individual
    precision_per_class, recall_per_class, f1_per_class, _ = precision_recall_fscore_support(
        labels, preds, average=None
    )

    metrics = {
        "accuracy": acc,
        "f1_macro": f1_macro,  # ✅ Métrica principal
        "precision_macro": precision_macro,
        "recall_macro": recall_macro,
        "f1_weighted": f1_weighted,
        "precision_weighted": precision_weighted,
        "recall_weighted": recall_weighted,
    }

    # Agregar F1 por clase
    for i in range(len(f1_per_class)):
        metrics[f"f1_class_{i}"] = f1_per_class[i]

    return metrics

# ==============================================
# ⚙️ 8. CONFIGURACIÓN DEL ENTRENAMIENTO
# ==============================================
wandb.init(
    project="commit-classification-roberta_V2",
    name="roberta-4clases-classweights-f1macro",
    config={
        "model": "roberta-base",
        "num_labels": 4,
        "learning_rate": 2e-5,
        "batch_size": 16,
        "epochs": 10,
        "max_length": 128,
        "balancing_strategy": "class_weights",
        "optimization_metric": "f1_macro",
        "class_weights": class_weights.tolist(),
        "weight_decay": 0.01
    }
)

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=10,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=50,
    load_best_model_at_end=True,
    metric_for_best_model="f1_macro",  # ✅ Optimizar F1 macro
    greater_is_better=True,
    dataloader_pin_memory=False,
    fp16=True,
    report_to="wandb",
    warmup_steps=100,
    save_total_limit=2
)

print("\n⚙️ Configuración del entrenamiento:")
print(f"   - Estrategia: Class Weights (sin oversampling)")
print(f"   - Métrica objetivo: F1-macro")
print(f"   - Learning rate: 2e-5")
print(f"   - Epochs: 5")
print(f"   - Batch size: 16")
print(f"   - Class weights aplicados en la loss function")

# ==============================================
# 🚀 9. ENTRENAMIENTO DEL MODELO
# ==============================================
trainer = WeightedLossTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
    class_weights=class_weights_tensor
)

print("\n🚀 Iniciando entrenamiento con class weights...\n")
print("   ⏱️ Tiempo estimado: ~15-20 minutos con GPU")
print("   📊 Sigue el progreso en tu dashboard de Wandb\n")
trainer.train()

# ==============================================
# 🧾 10. EVALUACIÓN FINAL
# ==============================================
metrics = trainer.evaluate()
print("\n📈 Resultados de evaluación:")
print(f"\n   🎯 Métricas principales:")
print(f"   - Accuracy: {metrics['eval_accuracy']:.4f}")
print(f"   - F1-macro: {metrics['eval_f1_macro']:.4f} ⭐")
print(f"   - F1-weighted: {metrics['eval_f1_weighted']:.4f}")
print(f"   - Precision-macro: {metrics['eval_precision_macro']:.4f}")
print(f"   - Recall-macro: {metrics['eval_recall_macro']:.4f}")

print(f"\n   📊 F1-score por clase:")
for i in range(4):
    print(f"   - Clase {i}: {metrics[f'eval_f1_class_{i}']:.4f}")

# ==============================================
# 📉 11. VISUALIZACIÓN DE MÉTRICAS
# ==============================================
preds_output = trainer.predict(tokenized_test)
y_true = preds_output.label_ids
y_pred = preds_output.predictions.argmax(-1)

# Matriz de confusión
cm = confusion_matrix(y_true, y_pred, labels=[0, 1, 2, 3])

wandb.log({
    "confusion_matrix": wandb.plot.confusion_matrix(
        probs=None,
        y_true=y_true,
        preds=y_pred,
        class_names=["0-Doc", "1-Amb", "2-NoWhat", "3-NoWhy"]
    )
})

plt.figure(figsize=(8,7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=[0, 1, 2, 3], yticklabels=[0, 1, 2, 3])
plt.xlabel("Predicciones")
plt.ylabel("Etiquetas reales")
plt.title("Matriz de Confusión - RoBERTa con Class Weights")
plt.tight_layout()
plt.show()

# Reporte detallado
labels_names = [
    "Documentado (Why+What)",
    "Ambiguo (No Why, No What)",
    "Vago - No What",
    "Vago - No Why"
]
report = classification_report(y_true, y_pred, target_names=labels_names, output_dict=True)
report_df = pd.DataFrame(report).transpose()

print("\n📋 Reporte detallado por clase:")
print(report_df)

# Gráfico comparativo F1-macro vs F1-weighted vs F1 por clase
plt.figure(figsize=(12,6))

# F1 por clase
clases = [0, 1, 2, 3]
f1_scores = [report[labels_names[i]]["f1-score"] for i in clases]
colors = ['#2E7D32', '#F57C00', '#C62828', '#6A1B9A']
nombres_cortos = ['Doc\n(0)', 'Amb\n(1)', 'NoWhat\n(2)', 'NoWhy\n(3)']

x_pos = np.arange(len(clases))
bars = plt.bar(x_pos, f1_scores, color=colors, alpha=0.7, edgecolor='black', label='F1 por clase')

# Línea de F1-macro
f1_macro_value = np.mean(f1_scores)
plt.axhline(y=f1_macro_value, color='red', linestyle='--', linewidth=2, label=f'F1-macro: {f1_macro_value:.3f}')

# Línea de F1-weighted
support = [report[labels_names[i]]["support"] for i in clases]
f1_weighted_value = np.average(f1_scores, weights=support)
plt.axhline(y=f1_weighted_value, color='blue', linestyle='--', linewidth=2, label=f'F1-weighted: {f1_weighted_value:.3f}')

plt.ylim(0, 1)
plt.xlabel("Clase")
plt.ylabel("F1-score")
plt.title("📊 Comparación F1: Por Clase vs Macro vs Weighted")
plt.xticks(x_pos, nombres_cortos)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Log métricas individuales a wandb
for i, nombre in enumerate(labels_names):
    wandb.log({
        f"f1_score_clase_{i}": f1_scores[i],
        f"precision_clase_{i}": report[nombre]["precision"],
        f"recall_clase_{i}": report[nombre]["recall"]
    })

# Análisis de errores
print("\n📊 Análisis de predicciones por clase:")
for i in range(4):
    mascara_real = y_true == i
    total_clase = np.sum(mascara_real)
    correctos = np.sum((y_true == i) & (y_pred == i))
    incorrectos = total_clase - correctos

    if total_clase > 0:
        precision_clase = (correctos / total_clase) * 100
        print(f"\nClase {i} - {labels_names[i]}:")
        print(f"   Total: {total_clase} ejemplos")
        print(f"   Correctos: {correctos} ({precision_clase:.1f}%)")
        print(f"   Incorrectos: {incorrectos} ({100-precision_clase:.1f}%)")
        print(f"   Class weight aplicado: {class_weights[i]:.4f}")

        if incorrectos > 0:
            print(f"   Confusiones:")
            for j in range(4):
                if i != j:
                    confusiones = np.sum((y_true == i) & (y_pred == j))
                    if confusiones > 0:
                        print(f"      - {confusiones} casos → Clase {j} ({labels_names[j]})")

# ==============================================
# 💾 12. GUARDAR Y DESCARGAR MODELO
# ==============================================
model.save_pretrained("./modelo_roberta_classweights_f1macro")
tokenizer.save_pretrained("./modelo_roberta_classweights_f1macro")

# Guardar configuración
config_info = {
    "num_labels": 4,
    "label_mapping": {
        "0": "Documentado (Why + What)",
        "1": "Ambiguo (Neither Why nor What)",
        "2": "No What",
        "3": "No Why"
    },
    "class_weights": class_weights.tolist(),
    "optimization_metric": "f1_macro",
    "training_strategy": "class_weights_in_loss"
}

import json
with open("./modelo_roberta_classweights_f1macro/model_config.json", "w") as f:
    json.dump(config_info, f, indent=2)

print("\n✅ Modelo guardado en './modelo_roberta_classweights_f1macro'")

# Comprimir y descargar
!zip -r modelo_roberta_classweights_f1macro.zip modelo_roberta_classweights_f1macro
files.download("modelo_roberta_classweights_f1macro.zip")

# Guardar métricas
metricas_por_clase = pd.DataFrame({
    "Clase": [0, 1, 2, 3],
    "Descripcion": labels_names,
    "Precision": [report[labels_names[i]]["precision"] for i in range(4)],
    "Recall": [report[labels_names[i]]["recall"] for i in range(4)],
    "F1-score": [report[labels_names[i]]["f1-score"] for i in range(4)],
    "Support": [report[labels_names[i]]["support"] for i in range(4)],
    "Class_Weight": class_weights.tolist()
})

metricas_por_clase.to_csv("metricas_classweights_f1macro.csv", index=False)
files.download("metricas_classweights_f1macro.csv")

# Resumen final
print("\n🎉 ¡Proceso completado!")
print(f"\n📊 Resumen de métricas:")
print(f"   - Accuracy: {metrics['eval_accuracy']:.4f}")
print(f"   - F1-macro (objetivo): {metrics['eval_f1_macro']:.4f} ⭐")
print(f"   - F1-weighted: {metrics['eval_f1_weighted']:.4f}")
print(f"\n⚖️ Estrategia utilizada:")
print(f"   - Class Weights en loss function")
print(f"   - Sin oversampling artificial")
print(f"   - Optimización basada en F1-macro")

wandb.finish()
```
